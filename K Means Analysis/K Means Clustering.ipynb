# Performing K Means Clustering

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

### Creating Own Artificial Dataset

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 200, n_features = 2, centers= 4, cluster_std= 1.8, random_state = 101)
```

make_blob return a tuple here with first element of tuple being the list of values,  
while the second element the labels of the cluster the element belongs to.  

```python
plt.scatter(data[0][:,0],data[0][:,1], c = data[1], cmap = 'viridis')
```

## Using KMeans to cluster the above created dataset.

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',max_iter = 300, n_init = 10)
    kmeans.fit(data[0])
    wcss.append(kmeans.inertia_) 
plt.plot(range(1,11),wcss)
plt.xticks(np.arange(1,11))
plt.yticks(np.arange(1,max(wcss),10000))
plt.grid()
plt.title('Within Clusters Squared Sum Visualisation')
plt.xlabel('No. of CLusters')
plt.ylabel('WCSS Sum')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

### Now Visualising

```python
fig , axes = plt.subplots(nrows = 1, ncols= 2, sharey =True, figsize = (10,6))

axes[0].set_title('Kmeans')
axes[0].scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_)

axes[1].set_title('Kmeans')
axes[1].scatter(data[0][:,0], data[0][:,1], c = data[1])
```

It can be observed that there is some noise around the borders of the clusters.
