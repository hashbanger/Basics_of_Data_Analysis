# Visualising with Pandas Built-ins

## -----------------------------------------------------------------------------------------------------------------

```python
import pandas as pd
```

Importing **seaborn** will automatically make the plots nicer!

```python
import seaborn as sns
```

```python
df1 = pd.read_csv('df1', index_col = 0)
```

```python
df1.head()
```

```python
df1['A'].hist()
```

As it can be seen that **pandas** is calling **matplotlib** under the hood.

```python
df1['A'].plot(kind = 'hist')
```

```python
df1['A'].plot.hist(bins = 30, color = 'red', alpha = 0.4)
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
df2.plot.area(alpha = 0.6 ,cmap = 'viridis')
```

```python
df2.plot.bar(stacked = True, cmap = 'Dark2')
```

It is good for categorical indexes.

```python
df1.plot.line(x = df1.index,y = 'B' ,figsize = (15,6))
```

```python
df1.plot.scatter(x = 'A', y = 'B', figsize = (8,8))
```

```python
df1.plot.scatter(x = 'A', y = 'B',c= 'C',cmap = 'magma', figsize = (8,8))
```

Usually people like to distinguish over the size rather than colour. So we can do as following:

```python
sns.set_style(style = 'dark')
df1.plot.scatter(x = 'A', y = 'B',s = df1['C']*100, figsize = (8,8))
```

```python
import numpy as np
df = pd.DataFrame(data = np.random.randn(1000,2), columns = ['a','b'] )
df.head()
```

```python
df.plot.hexbin(x = 'a', y= 'b', gridsize = 19, figsize = (10,7), cmap = 'viridis')
```

```python
df['a'].plot.kde()
```

```python
df['a'].plot.density()
```

```python
df2.plot.density()
```
